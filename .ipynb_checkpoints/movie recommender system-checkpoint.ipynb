```python
import ast
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L    
```

```python
# ast.literal_eval is a function that evaluate a string as a literal expression, and return the result
```

```python
movies_df["genres"].apply(convert)
```

```python
movies_df['genres'] = movies_df["genres"].apply(convert)
```

```python
movies_df.head(1)
```

```python
movies_df['keywords'] = movies_df['keywords'].apply(convert)
```

```python
movies_df.head(1)
```

```python
def convert3(obj):
    L=[]
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L 
```

```python
movies_df['cast'] = movies_df['cast'].apply(convert3)
```

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
        break    
    return L 
```

```python
movies_df['crew'] =movies_df['crew'].apply(fetch_director)
```

```python
movies_df['genres']=movies_df['genres'].apply(lambda x:[i.replace(" ","")for i in x])
movies_df['keywords']=movies_df['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
movies_df['cast']=movies_df['cast'].apply(lambda x:[i.replace(" ","")for i in x])
movies_df['crew']=movies_df['crew'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movies_df['tags'] = movies_df['overview']+ movies_df['genres']+movies_df['keywords']+ movies_df['cast']+ movies_df['crew']
```

```python
new_df = movies_df[["movie_id","title","tags"]]
```

```python
 new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags'] =  new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y =[]
    
    for i in text.split():
        y.append(ps.stem(i))
        
    return  " ". join(y)  
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
#bag of words means 

# The bag-of-words (BoW) model is a popular technique in natural language processing (NLP) that represents text data as an
# unordered collection of words. It disregards the order of words but captures their frequency, allowing for the analysis of
# text without considering grammar or syntax. The model consists of two main components: a vocabulary of known words and a
# measure of the presence of these words within a document. This approach is widely used in various applications, including 
# information retrieval and machine learning algorithms. 
```

```python
# stop words
# In Natural Language Processing (NLP), stop words are common words such as "the", "is", "and", and "in" that carry little 
# semantic meaning and are often removed to improve processing efficiency and model accuracy. They can be standard, 
# domain-specific, contextual, or even numerical depending on the dataset and application.
```
